```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# read data from dataset
dataset = pd.read_csv('konsumen.csv')
dataset.keys()
```

```python
mydata = pd.DataFrame(dataset)
mydata.head(10)
```

```python
#convert data to array
array_data = np.asarray(dataset)
print(array_data)
```

```python
#display data in a scater plot
plt.scatter(array_data[:, 0], array_data[:, 1], label='True Position')
plt.xlabel('Salary')
plt.ylabel('Spending')
plt.title('Consumer data distribution graph')
plt.show()
```

```python
#Activation K-means with K = 2
kmeans = KMeans(n_clusters=2)
kmeans.fit(array_data)
```

```python
#Displays the centroid value generated by the algorithm
print(kmeans.cluster_centers_)
```

```python
#Plot data point
#Visualise the results of consumer data clustering
plt.scatter(array_data[:,0], array_data[:,1], c=kmeans.labels_, cmap='rainbow')
plt.xlabel('Salary')
plt.ylabel('Spending')
plt.title('Consumer data distribution graph')
plt.show()
```

```python
#Plot data point
#Visualise the clustering results with the centroid of each cluster
plt.scatter(array_data[:,0], array_data[:,1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color='black')
plt.xlabel('Salary')
plt.ylabel('Spending')
plt.title('Consumer data distribution graph')
plt.show()
```
