```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

numpy => komputasi matrix
pandas => import file
matplotlib => visualisai

```python
dataset = pd.read_csv("Social_Network_Ads.csv")
x = dataset.iloc[:, [2,3]].values
y  = dataset.iloc[:, -1].values
```

```python
print(x)
print(y)
```

x => megambil data age, dan estimation 

y => mengambil data class

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

```python
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

membuat data x, dan y menjadi data test dan data test

75% => data train

25% => data test

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
print(x_train)
```

```python
print(x_test)
```

mengecilkan skala data x_train dan x_test

```python
from sklearn.naive_bayes import GaussianNB # menggunakan gaussiantNB karena data numerik
classifier = GaussianNB()
classifier.fit(x_train, y_train)
```

```python
# membuat prediksi dari data training
y_prediction = classifier.predict(x_test)
```

```python
# menghitung akurasi <confusion matrix> untuk efaluasi, dari data testing terhadap data preduction
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_prediction)
print(cm)
```

cara menghitung prediction adalah tambahkan semua data pada matrix 65+3+7+25 dibagi 65 + 25 = 100/90 = 0.9

jadi akurasi 90%

```python
# mgengambarkan persebaran data
from matplotlib.colors import ListedColormap
x_set, y_set = x_train, y_train
x1, x2 = np.meshgrid(np.arange(start = x_set[:, 0].min()-1, stop = x_set[:, 0].max()+1, step=0.01),
                     np.arange(start = x_set[:, 1].min()-1, stop = x_set[:, 0].max()+1, step=0.01))
plt.contourf(x1, x2, classifier.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))

plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())

for i, j in enumerate (np.unique(y_set)):
  plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1],
              c = ListedColormap(('red', 'green'))(i), label = j)
  
plt.title('Data classfication [data training] with Naive Bayes')
plt.xlabel('Age')
plt.ylabel('Salary Estimation')
plt.legend()
plt.show()
```

```python
# mgengambarkan persebaran data
from matplotlib.colors import ListedColormap
x_set, y_set = x_test, y_test
x1, x2 = np.meshgrid(np.arange(start = x_set[:, 0].min()-1, stop = x_set[:, 0].max()+1, step=0.01),
                     np.arange(start = x_set[:, 1].min()-1, stop = x_set[:, 0].max()+1, step=0.01))
plt.contourf(x1, x2, classifier.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))

plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())

for i, j in enumerate (np.unique(y_set)):
  plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1],
              c = ListedColormap(('red', 'green'))(i), label = j)
  
plt.title('Data classfication [data testing] with Naive Bayes')
plt.xlabel('Age')
plt.ylabel('Salary Estimation')
plt.legend()
plt.show()
```
