```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('top-anime.csv', encoding='latin-1')
dataset.head()
```

```python
selected_column = dataset.iloc[:, [4, 14, 15, 17, 18]]
print(selected_column)
```

```python
print(selected_column.dtypes)
```

```python
for column_index in range(len(selected_column.columns)):
    if selected_column.dtypes[column_index] == 'object':
        selected_column.iloc[:, column_index] = selected_column.iloc[:, column_index].str.replace(',', '')
        selected_column.iloc[:, column_index] = selected_column.iloc[:, column_index].astype(int)
```

```python
selected_column.head()
```

```python
selected_column = selected_column.iloc[1:]
```

```python
import scipy.cluster.hierarchy as sch
```

```python
dendrogram = sch.dendrogram(sch.linkage(selected_column, method = 'ward',  metric='euclidean'))
plt.title('Dendrogram')
plt.xlabel('Top Anime')
plt.ylabel('Euclidean Distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac = AgglomerativeClustering(n_clusters = 6, affinity = 'euclidean', linkage = 'ward')
predicted = ac.fit_predict(selected_column)
```

```python
plt.scatter(selected_column[predicted == 0].iloc[:, 0], selected_column[predicted == 0].iloc[:, 1], s=100, c='red', label='Cluster 1')
plt.scatter(selected_column[predicted == 1].iloc[:, 0], selected_column[predicted == 1].iloc[:, 1], s=100, c='blue', label='Cluster 2')
plt.scatter(selected_column[predicted == 2].iloc[:, 0], selected_column[predicted == 2].iloc[:, 1], s=100, c='green', label='Cluster 3')
plt.scatter(selected_column[predicted == 3].iloc[:, 0], selected_column[predicted == 3].iloc[:, 1], s=100, c='cyan', label='Cluster 4')
plt.scatter(selected_column[predicted == 4].iloc[:, 0], selected_column[predicted == 4].iloc[:, 1], s=100, c='magenta', label='Cluster 5')
plt.scatter(selected_column[predicted == 5].iloc[:, 0], selected_column[predicted == 5].iloc[:, 1], s=100, c='purple', label='Cluster 6')
plt.title('Clusters of top anime')
plt.legend()
plt.show()

```
